### ① 取出今天所有的發文

```python
from bs4 import BeautifulSoup
import requests
url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)
soup = BeautifulSoup(r.text, "html")
for d in soup.find(class_="part_list_2").findChildren('h3'):
    print(d.find(class_="date").text, d.find_all('a')[0].text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python

dic = {}
for d in soup.find(class_="part_list_2").findChildren('h3'):
#     print(d)
    date = d.find(class_="date").text
    cat = d.select(".tag")[0].text
    news = d.find_all('a')[0].text
    dic.setdefault(cat,[])
    dic[cat].append({
        'date':date,
        'news':news
    })
    
print(dic)
```

### ③ 哪一個類別的文章最多

```python
max_l = 0
max_k = ""
for key, l in dic.items():
    if len(l) > max_l:
        max_k = key
        max_l = len(l)
#     print(key, l)
print(max_k)
```
